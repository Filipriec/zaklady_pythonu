## A

Načítajte a, b a vypíš a+b, a-b, a*b, a/b (ak b sa nerovna 0, inak vypíšte hlášku (!=)).

```python
a = float(input("Zadaj a: "))
b = float(input("Zadaj b: "))

print("a+b =", a + b)
print("a-b =", a - b)
print("a*b =", a * b)

if b != 0:
    print("a/b =", a / b)
else:
    print("a/b: delenie nulou nie je mozne")
```

Načítajte celé n. Vypíšte „párne“ alebo „nepárne“.

```python
n = int(input("Zadaj cele cislo n: "))
if n % 2 == 0:
    print("parne")
else:
    print("neparne")
```

Načítajte a, b, c. Vypíšte najväčšie. (Bez max().)

```python
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))

m = a
if b > m:
    m = b
if c > m:
    m = c

print("max =", m)
```

Nacitajte n. Vypiste jeho absolutnu hodnotu.

```python
x = float(input("x: "))
if x < 0:
    x = -x
print("abs =", x)
```

```python
n = int(input("n: "))
if n < 0:
    print("negativne")
elif n == 0:
    print("nula")
else:
    print("pozitivne")
```

## B

Načítajte N. Spočítajte súčet 1+2+...+N pomocou cyklu.

```python
n = int(input("N: "))
s = 0
for i in range(1, n + 1):
    s += i
print("sucet =", s)
```

Načítajte N (0-12). Vypočítajte N! cyklom.

```python
n = int(input("N (0-12): "))
f = 1
for i in range(2, n + 1):
    f *= i
print("faktorial =", f)
```

Načítajte n > 0. Zistite, koľko má deliteľov.

```python
n = int(input("n > 0: "))
cnt = 0
for i in range(1, n + 1):
    if n % i == 0:
        cnt += 1
print("pocet delitelov =", cnt)
```

## C


Načítajte dva zoznamy čísel. Vypíšte spoločné prvky (každý len raz).

```python
a = [1, 2, 3, 4, 4, 7]
b = [0, 2, 4, 4, 6, 7, 7]

res = []
for x in a:
    if x in b and x not in res:
        res.append(x)

print("a =", a)
print("b =", b)
print("prienik =", res)
```

Alebo druhy sposob, ako sa da uloha vyriesit:

```python
a = [1, 2, 3, 4, 4, 7]
b = [0, 2, 4, 4, 6, 7, 7]
res = []
i = 0
j = 0

while i < len(a) and j < len(b):
    if a[i] == b[j]:
        # Overenie duplicity vo výsledku
        if len(res) == 0 or a[i] != res[len(res) - 1]:
            res.append(a[i])
        i += 1
        j += 1
    elif a[i] < b[j]:
        i += 1
    else:
        j += 1
```
