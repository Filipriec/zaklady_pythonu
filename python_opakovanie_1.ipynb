## 1. Premenn√© a priradenie

Premenn√° je **meno**, pod ktor√Ωm si poƒç√≠taƒç zapam√§t√° nejak√∫ hodnotu.

V Pythone nepotrebujeme hovori≈•, ak√Ω typ hodnoty do premennej uklad√°me ‚Äì Python si to zist√≠ s√°m.

````python
# Cel√© ƒç√≠slo (int)
x = 5
print(x)

# Desatinn√© ƒç√≠slo (float)
y = 3.14
print(y)

# Z√°porn√© ƒç√≠slo
z = -10
print(z)
````

Funkcia `type()` n√°m prezrad√≠, ak√©ho typu je hodnota v premennej:

````python
x = 5
print(type(x))   # int = cel√© ƒç√≠slo

y = 3.14
print(type(y))   # float = desatinn√© ƒç√≠slo
````

### 1.1 Zmena hodnoty premennej

Hodnotu premennej m√¥≈æeme kedykoƒævek **prep√≠sa≈•** ‚Äì star√° hodnota sa zabudne a nahrad√≠ novou.

````python
x = 10
print("x je", x)

x = 20            # prep√≠≈°eme star√∫ hodnotu
print("x je teraz", x)
````

Premenn√∫ m√¥≈æeme aj **meni≈• na z√°klade jej aktu√°lnej hodnoty**:

````python
x = 20
x = x + 5         # vezmi aktu√°lne x (20), priƒç√≠taj 5, ulo≈æ sp√§≈•
print("x + 5 =", x)

y = 20
y = y - 8
print("y - 8 =", y)

z = 4
z = z * 2
print("z * 2 =", z)

a = 3
a = a ** 2         # ** znamen√° umocnenie
print("a ** 2 =", a)
````

### 1.2 Skr√°ten√Ω z√°pis oper√°ci√≠

Z√°pis `x = x + 5` sa d√° skr√°ti≈• na `x += 5`. Funguje to rovnako pre v≈°etky oper√°cie:

| Dlh√Ω z√°pis | Skr√°ten√Ω z√°pis |
|---|---|
| `x = x + 3` | `x += 3` |
| `x = x - 3` | `x -= 3` |
| `x = x * 5` | `x *= 5` |
| `x = x / 5` | `x /= 5` |
| `x = x ** 2` | `x **= 2` |
| `x = x % 2` | `x %= 2` |

> ‚ö†Ô∏è Python **nem√°** oper√°tory `x++` ani `x--`, ktor√© pozn√°te z in√Ωch jazykov. Namiesto nich pou≈æite `x += 1` a `x -= 1`.

````python
x = 10
x += 3
print("po += 3:", x)

x -= 5
print("po -= 5:", x)

x *= 4
print("po *= 4:", x)

x /= 2
print("po /= 2:", x)
````

## 3. Z√°kladn√© d√°tov√© typy

### 3.1 Cel√© ƒç√≠sla (`int`) a desatinn√© ƒç√≠sla (`float`)

| Oper√°cia | Z√°pis | Pr√≠klad | V√Ωsledok |
|---|---|---|---|
| Sƒç√≠tanie | `+` | `3 + 2` | `5` |
| Odƒç√≠tanie | `-` | `3 - 2` | `1` |
| N√°sobenie | `*` | `3 * 2` | `6` |
| Delenie | `/` | `7 / 2` | `3.5` |
| Celoƒç√≠seln√© delenie | `//` | `7 // 2` | `3` |
| Modulo (zvy≈°ok) | `%` | `7 % 2` | `1` |
| Umocnenie | `**` | `3 ** 2` | `9` |

````python
# Vysk√∫≈°ajte si rozdiely:
print("7 / 2  =", 7 / 2)    # be≈æn√© delenie ‚Üí v≈ædy float
print("7 // 2 =", 7 // 2)   # celoƒç√≠seln√© delenie ‚Üí zahod√≠ zvy≈°ok
print("7 % 2  =", 7 % 2)    # modulo ‚Üí vr√°ti len zvy≈°ok
````

### 3.2 Boolovsk√© hodnoty (`bool`)

Boolovsk√© (logick√©) premenn√© maj√∫ iba dve hodnoty: `True` (pravda) a `False` (nepravda). Vznikaj√∫ napr√≠klad pri porovn√°van√≠:

````python
print(5 > 3)     # True
print(5 == 3)    # False
print(5 != 3)    # True  (nie je rovn√©)
````

````python
# Logick√© oper√°cie
t = True
f = False

print("t and f =", t and f)     # True, len ak s√∫ OBE pravdiv√©
print("t or f  =", t or f)      # True, ak je ASPO≈á JEDNA pravdiv√°
print("not t   =", not t)       # opak (neg√°cia)
````

## 7.1 ƒåo je funkcia a preƒço ju potrebujeme?

Predstavte si, ≈æe mus√≠te 10-kr√°t vypoƒç√≠ta≈• obvod obdƒ∫≈ænika pre r√¥zne rozmery. Bez funkcie by ste museli 10-kr√°t p√≠sa≈• ten ist√Ω vzorec. S funkciou ho nap√≠≈°ete **raz** a potom len vol√°te.






````python
# BEZ funkcie ‚Äì opakujeme sa:
print(2 * (5 + 3))
print(2 * (10 + 4))
print(2 * (7 + 2))

# S funkciou ‚Äì nap√≠≈°eme raz, vol√°me koƒækokr√°t chceme:
def obvod(a, b):
    return 2 * (a + b)

print(obvod(5, 3))
print(obvod(10, 4))
print(obvod(7, 2))
````

## 7.6 `return` vs `print` ‚Äì d√¥le≈æit√Ω rozdiel!

Toto je vec, kde sa ≈æiaci najƒçastej≈°ie pom√Ωlia:


````python
# Funkcia s RETURN ‚Äì vr√°ti hodnotu, m√¥≈æeme s ≈àou ƒèalej poƒç√≠ta≈•
def scitaj(a, b):
    return a + b

x = scitaj(3, 4)       # x je teraz 7
print(x * 2)           # 14 ‚Äì m√¥≈æeme s v√Ωsledkom pracova≈•

# Funkcia s PRINT ‚Äì iba vyp√≠≈°e, ale niƒç nevr√°ti
def scitaj_print(a, b):
    print(a + b)

y = scitaj_print(3, 4)  # vyp√≠≈°e 7, ale...
print(y)                 # None! Funkcia niƒç nevr√°tila
````

> **Pravidlo:** Ak chcete s v√Ωsledkom ƒèalej pracova≈•, pou≈æite `return`. Ak chcete len nieƒço zobrazi≈•, pou≈æite `print`.

````python
def je_parne(cislo):
    if cislo % 2 == 0:
        return True
    else:
        return False

print(je_parne(4))    # True
print(je_parne(7))    # False
````
