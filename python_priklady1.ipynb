## A

Načítajte a, b a vypíš a+b, a-b, a*b, a/b (ak b sa nerovna 0, inak vypíšte hlášku (!=)).

```python
a = float(input("Zadaj a: "))
b = float(input("Zadaj b: "))

print("a+b =", a + b)
print("a-b =", a - b)
print("a*b =", a * b)

if b != 0:
    print("a/b =", a / b)
else:
    print("a/b: delenie nulou nie je mozne")
```

Načítajte a, b, c. Vypíšte najväčšie. (Bez max().)

```python
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))

m = a
if b > m:
    m = b
if c > m:
    m = c

print("max =", m)
```

## B

Načítajte N. Spočítajte súčet 1+2+...+N pomocou cyklu.

```python
n = int(input("N: "))
s = 0
for i in range(1, n + 1):
    s += i
print("sucet =", s)
```

Načítajte N (0-12). Vypočítajte N! cyklom.

```python
n = int(input("N (0-12): "))
f = 1
for i in range(2, n + 1):
    f *= i
print("faktorial =", f)
```

Načítajte n > 0. Zistite, koľko má deliteľov.

```python
n = int(input("n > 0: "))
cnt = 0
for i in range(1, n + 1):
    if n % i == 0:
        cnt += 1
print("pocet delitelov =", cnt)
```

Načítajte n. Určite, či je prvočíslo. (Optimalizácia len do i*i <= n.)

```python
n = int(input("n: "))

if n < 2:
    print("nie je prvocislo")
else:
    i = 2
    ok = True
    while i * i <= n:
        if n % i == 0:
            ok = False
            break
        i += 1
    print("prvocislo" if ok else "nie je prvocislo")
```

Načítavajte čísla, kým nepríde 0. Potom vypíšte maximum zo zadaných (0 nepočítať).

```python
mx_set = False
mx = 0

while True:
    x = int(input("Zadaj cislo (0 koniec): "))
    if x == 0:
        break
    if (not mx_set) or x > mx:
        mx = x
        mx_set = True

if mx_set:
    print("max =", mx)
else:
    print("nezadal si ziadne cislo")
```

Načítavajte čísla, kým nepríde 0. Vypíšte priemer (ak bolo aspoň jedno číslo).

```python
s = 0
cnt = 0

while True:
    x = int(input("Zadaj cislo (0 koniec): "))
    if x == 0:
        break
    s += x
    cnt += 1

if cnt > 0:
    print("priemer =", s / cnt)
else:
    print("nezadal si ziadne cislo")
```

## C

Načítajte zoznam čísel. Vytvorte nový zoznam bez duplicit, ale v pôvodnom poradí.

```python
a = [1, 2, 2, 3, 1, 4, 4, 5]

b = []
for x in a:
    if x not in b:
        b.append(x)

print("a =", a)
print("bez duplicit =", b)
```

Načítajte dva zoznamy čísel. Vypíšte spoločné prvky (každý len raz).

```python
a = [1, 2, 3, 4, 4, 7]
b = [0, 2, 4, 4, 6, 7, 7]

res = []
for x in a:
    if x in b and x not in res:
        res.append(x)

print("a =", a)
print("b =", b)
print("prienik =", res)
```

Načítajte zoznam a číslo x. Nájdite index prvého výskytu x alebo vypíš -1.

```python
a = [5, 10, 7, 10, 3]
x = 10

idx = -1
for i in range(len(a)):
    if a[i] == x:
        idx = i
        break

print("a =", a)
print("x =", x)
print("index =", idx)
```
