## 2. Funkcia `range()` ‚Äì gener√°tor ƒç√≠sel

`range()` je pomocn√Ω objekt, ktor√Ω **generuje postupnos≈• cel√Ωch ƒç√≠sel**. S√°m o sebe niƒç nevyp√≠≈°e ‚Äì len priprav√≠ ƒç√≠sla, ktor√© potom pou≈æijeme v cykle.

Aby sme si mohli pozrie≈•, ak√© ƒç√≠sla `range()` generuje, prevedieme ho na zoznam pomocou `list()`.

### 2.1 `range(koniec)` ‚Äì jeden argument

Generuje ƒç√≠sla **od 0** po `koniec` (koniec **nie je zahrnut√Ω**).

````python
print(list(range(5)))    # 0, 1, 2, 3, 4  (NIE 5!)
print(list(range(10)))   # 0, 1, 2, ..., 9
````

> ‚ö†Ô∏è **D√¥le≈æit√©:** `range(5)` generuje 5 ƒç√≠sel (0, 1, 2, 3, 4), ale ƒç√≠slo 5 tam **nie je**! V≈ædy konƒç√≠ o jedno sk√¥r.

### 2.2 `range(zaƒçiatok, koniec)` ‚Äì dva argumenty

Ak nechcete zaƒç√≠na≈• od 0, zad√°te aj poƒçiatoƒçn√© ƒç√≠slo.

````python
print(list(range(1, 6)))    # 1, 2, 3, 4, 5
print(list(range(2, 10)))   # 2, 3, 4, 5, 6, 7, 8, 9
print(list(range(10, 15)))  # 10, 11, 12, 13, 14
````

### 2.3 `range(zaƒçiatok, koniec, krok)` ‚Äì tri argumenty

Tret√≠m argumentom men√≠te **veƒækos≈• kroku** (predvolen√Ω krok je 1).

````python
print(list(range(0, 10, 2)))   # p√°rne: 0, 2, 4, 6, 8
print(list(range(1, 10, 2)))   # nep√°rne: 1, 3, 5, 7, 9
print(list(range(0, 50, 10)))  # desiatky: 0, 10, 20, 30, 40
print(list(range(2, 10, 3)))   # krok 3: 2, 5, 8
````

### 2.4 `range` s **z√°porn√Ωm** krokom ‚Äì odpoƒç√≠tavanie

Ak chcete √≠s≈• **smerom dole**, krok mus√≠ by≈• z√°porn√Ω a zaƒçiatok mus√≠ by≈• v√§ƒç≈°√≠ ako koniec.

````python
print(list(range(10, 0, -1)))   # 10, 9, 8, ..., 1
print(list(range(5, -1, -1)))   # 5, 4, 3, 2, 1, 0
print(list(range(20, 0, -5)))   # 20, 15, 10, 5
````

### Prehƒæad `range()`

| Z√°pis | V√Ωsledok | Vysvetlenie |

| `range(5)` | 0, 1, 2, 3, 4 | od 0, krok 1 |

| `range(10, 0, -1)` | 10, 9, ..., 1 | odpoƒç√≠tavanie |

## 3. Cyklus `for`

Cyklus `for` vykon√° blok k√≥du **pre ka≈æd√Ω prvok** v nejakej postupnosti.

```
for premenna in postupnost:
    # telo cyklu (odsaden√©!)
    # vykon√° sa raz pre ka≈æd√Ω prvok
```

V≈°imnite si:
- **dvojbodku** `:` za postupnos≈•ou
- telo cyklu je **odsaden√©** 4 medzerami (rovnako ako pri `if`)

### 3.1 Najjednoduch≈°√≠ `for` cyklus

````python
# Vyp√≠≈°e ƒç√≠sla 0 a≈æ 4
for i in range(5):
    print(i)
````

ƒåo sa tu deje krok po kroku:
1. `range(5)` vytvor√≠ postupnos≈•: 0, 1, 2, 3, 4
2. Prv√Ω priebeh: `i = 0`, vykon√° sa `print(0)`
3. Druh√Ω priebeh: `i = 1`, vykon√° sa `print(1)`
4. ... a tak ƒèalej, a≈æ k√Ωm sa nevyƒçerpaj√∫ v≈°etky ƒç√≠sla.

### 3.2 `for` s r√¥znymi variantmi `range()`

````python
# ƒå√≠sla od 1 do 5
print("ƒå√≠sla 1 a≈æ 5:")
for i in range(1, 6):
    print(i)
````

````python
# P√°rne ƒç√≠sla od 2 do 10
print("P√°rne ƒç√≠sla:")
for i in range(2, 11, 2):
    print(i)
````

````python
# Odpoƒç√≠tavanie
print("Odpoƒç√≠tavanie:")
for i in range(5, 0, -1):
    print(i)
print("≈†tart!")
````

### 3.3 V tele cyklu m√¥≈æeme robi≈• ƒçokoƒævek

Telo cyklu nie je obmedzen√© na `print`. M√¥≈æeme tam poƒç√≠ta≈•, pou≈æ√≠va≈• podmienky, vola≈• funkcie...

````python
# V√Ωpoƒçet druh√Ωch mocn√≠n
for i in range(1, 6):
    print(f"{i} na druh√∫ je {i ** 2}")
````

````python
# Podmienka vn√∫tri cyklu ‚Äì ktor√© ƒç√≠sla s√∫ p√°rne a ktor√© nep√°rne?
for i in range(1, 11):
    if i % 2 == 0:
        print(f"{i} je p√°rne")
    else:
        print(f"{i} je nep√°rne")
````

### 3.4 Sƒç√≠tavanie v cykle ‚Äì akumul√°tor

Veƒæmi ƒçast√Ω vzor: chceme **spoƒç√≠ta≈• s√∫ƒçet** viacer√Ωch ƒç√≠sel. Na to si vytvor√≠me premenn√∫ (akumul√°tor), ktor√∫ v ka≈ædom priechode cyklu zv√Ω≈°ime.

````python
# S√∫ƒçet ƒç√≠sel od 1 do 10
sucet = 0                     # akumul√°tor ‚Äì zaƒç√≠name na nule

for i in range(1, 11):
    sucet += i                # v ka≈ædom priechode pripoƒç√≠tame i
    print(f"Priechod {i}: sucet = {sucet}")

print(f"\nV√Ωsledn√Ω s√∫ƒçet 1 + 2 + ... + 10 = {sucet}")
````

### 3.5 `for` cez re≈•azec

Cyklus `for` v Pythone **nie je obmedzen√Ω na ƒç√≠sla**. M√¥≈æete iterova≈• cez ƒçokoƒævek ‚Äì napr√≠klad cez p√≠smen√° v re≈•azci:

````python
slovo = "Python"

for pismeno in slovo:
    print(pismeno)
````

## 4. Cyklus `while`

Cyklus `while` opakuje blok k√≥du **dok√Ωm je podmienka pravdiv√°** (`True`).

```
while podmienka:
    # telo cyklu (odsaden√©!)
    # vykon√°va sa, k√Ωm je podmienka True
```

### Kedy `while` a kedy `for`?

| Situ√°cia | Pou≈æite |
|---|---|
| Viem **koƒækokr√°t** chcem opakova≈• | `for` |
| Chcem prejs≈• cez postupnos≈• (ƒç√≠sla, p√≠smen√°...) | `for` |
| Neviem koƒækokr√°t ‚Äì z√°vis√≠ to od **podmienky** | `while` |
| Chcem opakova≈• **dok√Ωm** nieƒço plat√≠ | `while` |

### 4.1 Najjednoduch≈°√≠ `while` cyklus

````python
# Vyp√≠≈°e ƒç√≠sla 0 a≈æ 4
i = 0

while i < 5:
    print(i)
    i += 1       # D√îLE≈ΩIT√â: bez tohto by cyklus be≈æal donekoneƒçna!
````

ƒåo sa tu deje krok po kroku:
1. `i = 0` ‚Äì nastav√≠me poƒçiatoƒçn√∫ hodnotu
2. Je `i < 5`? √Åno (0 < 5) ‚Üí vykon√° sa telo: vyp√≠≈°e 0, zv√Ω≈°i `i` na 1
3. Je `i < 5`? √Åno (1 < 5) ‚Üí vykon√° sa telo: vyp√≠≈°e 1, zv√Ω≈°i `i` na 2
4. ... pokraƒçuje ...
5. Je `i < 5`? Nie (5 < 5 je False) ‚Üí cyklus **konƒç√≠**

> ‚ö†Ô∏è **Pozor na nekoneƒçn√Ω cyklus!** Ak zabudnete meni≈• premenn√∫ v podmienke (napr. `i += 1`), cyklus sa nikdy nezastav√≠. Ak sa v√°m to stane, stlaƒçte tlaƒçidlo **zastavenia** vedƒæa bunky (‚¨ú).

### 4.2 `while` ‚Äì keƒè nevieme koƒækokr√°t

Tu uvid√≠te silu `while` ‚Äì ned√° sa jednoducho nahradi≈• `for` cyklom:

````python
# Opakovan√© umoc≈àovanie: zaƒç√≠name na 2, umoc≈àujeme na druh√∫,
# k√Ωm neprekroƒç√≠me 100
c = 2

while c < 100:
    c **= 2
    print(c)
````

Na zaƒçiatku sme nevedeli, koƒæko priebehov to bude ‚Äì vedeli sme len **podmienku**, kedy presta≈•.

````python
# Koƒækokr√°t mus√≠me vydeli≈• ƒç√≠slo dvoma, k√Ωm dostaneme menej ako 1?
cislo = 1000
pocet = 0

while cislo >= 1:
    cislo /= 2
    pocet += 1

print(f"Delili sme {pocet}-kr√°t, zostalo {cislo}")
````

### 4.3 `while` s podmienkou a vstupom od pou≈æ√≠vateƒæa

Typick√© pou≈æitie `while`: opakujeme nieƒço, **k√Ωm pou≈æ√≠vateƒæ neskonƒç√≠**.

````python
# Sk√∫ste spusti≈• a zad√°va≈• ƒç√≠sla. Zadajte 0 pre ukonƒçenie.
sucet = 0
cislo = int(input("Zadajte ƒç√≠slo (0 = koniec): "))

while cislo != 0:
    sucet += cislo
    cislo = int(input("Zadajte ƒç√≠slo (0 = koniec): "))

print(f"S√∫ƒçet zadan√Ωch ƒç√≠sel je {sucet}")
````

## 5. Porovnanie `for` a `while`

Ten ist√Ω v√Ωstup sa ƒçasto d√° dosiahnu≈• oboma sp√¥sobmi. Pozrite sa na rozdiel:

````python
# V√Ωpis ƒç√≠sel 0 a≈æ 4 pomocou FOR
print("=== FOR ===")
for i in range(5):
    print(i)

# Ten ist√Ω v√Ωstup pomocou WHILE
print("=== WHILE ===")
i = 0
while i < 5:
    print(i)
    i += 1
````

V≈°imnite si, ≈æe `for` verzia je **krat≈°ia a prehƒæadnej≈°ia** ‚Äì nemus√≠me ruƒçne inicializova≈• ani zvy≈°ova≈• `i`. Preto ak viete koƒækokr√°t chcete opakova≈•, **v≈ædy preferujte `for`**.

## 6. `break` a `continue` ‚Äì riadenie cyklu

Niekedy chcete cyklus **predƒçasne ukonƒçi≈•** alebo **preskoƒçi≈•** jeden priechod.

### 6.1 `break` ‚Äì okam≈æit√© ukonƒçenie cyklu

````python
# Hƒæad√°me prv√© ƒç√≠slo deliteƒæn√© 7 v rozsahu 1-100
for i in range(1, 101):
    if i % 7 == 0:
        print(f"Prv√© ƒç√≠slo deliteƒæn√© 7 je {i}")
        break    # na≈°li sme, konƒç√≠me cyklus
````

### 6.2 `continue` ‚Äì preskoƒç zvy≈°ok a pokraƒçuj ƒèal≈°√≠m priebehom

````python
# Vyp√≠≈° ƒç√≠sla 1-10, ale preskoƒç ƒç√≠slo 5
for i in range(1, 11):
    if i == 5:
        continue    # preskoƒç√≠ print a ide na ƒèal≈°ie i
    print(i)
````

## 7. Cykly a funkcie ‚Äì kombin√°cia

Cykly m√¥≈æeme umiestni≈• do funkci√≠ a naopak ‚Äì vo funkci√°ch vola≈• cykly. To je z√°klad programovania.

````python
def sucet_od_do(od, do):
    """Spoƒç√≠ta s√∫ƒçet cel√Ωch ƒç√≠sel od 'od' po 'do' (vr√°tane)."""
    vysledok = 0
    for i in range(od, do + 1):
        vysledok += i
    return vysledok

print(sucet_od_do(1, 10))    # 55
print(sucet_od_do(1, 100))   # 5050
````

````python
def je_prvocislo(n):
    """Zist√≠, ƒçi je ƒç√≠slo n prvoƒç√≠slo."""
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False     # na≈°li sme deliteƒæa ‚Üí nie je prvoƒç√≠slo
    return True              # nena≈°li sme deliteƒæa ‚Üí je prvoƒç√≠slo

# Test
for cislo in range(1, 21):
    if je_prvocislo(cislo):
        print(f"{cislo} je prvoƒç√≠slo")
````
